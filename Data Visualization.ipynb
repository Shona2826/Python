<h1>Data Visualization</h1><br>A fundamental part of the data scientist's toolkit is data visualization. The two primary uses of data visualizationa are:<br>1. To <b>Explore</b> data <br> 2. To <b>Communicate</b> data

<h3>Matplotlib</h3><br> Here we will import matplotlib.pyplot module .pyplot maintains internal state in which you build up a visualization step by step. It works pretty well for simple bar chats,line chats and scatterplots.

```python
from matplotlib import pyplot as plt 
years = [2001,2002,2003,2004,2005,2006]
gdp = [300,340,350,500,700,890]

plt.plot(years,gdp,color = 'red', marker='o',linestyle = 'solid')
plt.title("Nominla GDP")
plt.ylabel("Billions of $")
plt.show()
```

<h3>Bar charts</h3> <br> Its a good choice when we want to show how some quantity varies among some discrete set of items.

```python
#bars by default width is 0.8 
#lets plot the above data using barplot
plt.bar(years,gdp)
plt.title("Nominla GDP")
plt.ylabel("Billions of $")
```

A bar chart can also be a good choice for plotting histograms of bucketed numeric values, in order to visually explore how the values are distributed

```python
from collections import Counter
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter( decile( grade) for grade in grades)

plt. bar( [x - 4 for x in histogram. keys( ) ] , # shift each bar to the left by 4
histogram. values( ) , # give each bar its correct height
8) # give each bar a width of 8
plt. xlabel( "Decile")
plt. ylabel( "No. of Students")
plt. title( "Distribution of Exam 1 Grades")
plt. show( )
```

<h3> Line chart</h3>It's a good choice for showing trends

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip( variance, bias_squared) ]
xs = [i for i, _ in enumerate( variance) ]
```

We can call multiple calls to plt.plot to show multiple series on the same chart

```python

plt.plot(xs, variance, 'g-', label='variance') # Here 'g-' means green line plots
plt.plot(xs, bias_squared,'b-.' , label='bias sqared') 
plt.plot(xs, total_error, 'r:', label='total error') # 'r:' means red dotted plots
#loc =9 means top center
plt. legend( loc=9)
plt. xlabel( "model complexity")
plt. title( "The Bias- Variance Tradeoff")
plt. show( )
```

<h3>Scatterplots</h3>It is a right choice for visualizing realtionship between two paired sets of data.

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = [' a' , ' b' , ' c' , ' d' , ' e' , ' f' , ' g' , ' h' , ' i' ]
plt. scatter( friends, minutes)
# label each point
for label, friend_count, minute_count in zip( labels, friends, minutes) :
    plt. annotate( label,
        xy=( friend_count, minute_count) , # put the label with its point
        xytext=( 5, - 5) , # but slightly offset
        textcoords=' offset points' )
plt. title( "Daily Minutes vs. Number of Friends")
plt. xlabel( "No. of friends")
plt. ylabel( "Daily minutes spent on the site")
plt. show( )
```
